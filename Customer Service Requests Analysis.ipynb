```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.info()
```

```python
#calculating the Request_Closing_Time in terms of hours
def req_closing_time(df):
    df['request_closing_time']=((df['closed_date']-df['created_date']).dt.seconds/60)/60
    df1=df[df['request_closing_time'].notnull()] #avoiding not null
    df_close_time=df1[df1['closed_date']>=df1['created_date']]
    return df_close_time

df_new=req_closing_time(df)
df_new
```

```python
# doing statistical test to accept or reject the Null Hypothesis along with the corresponding ‘p-value’
```

```python
#taking sample n=3000 from df_new data set

df_sample=df_new.sample(n=3000)
df_sample
```

```python
#finding average response time for request_closing_time
avg_res_time= df_new['request_closing_time'].mean()
avg_res_time
```

```python
#cheking the shape of sample
df_sample.shape
```

```python
#Shapiro-Wilk Test of Normality
#H0: request_closing_time is normal

from scipy.stats import shapiro
stat,p= shapiro(df_sample.request_closing_time)

print(stat, p)
```

As P-value=0.0 which is less than 0.05, we can reject the null hypothesis and there is evidence that data is not normally distributed population.

```python
#lets check for QQPlot

import statsmodels.api as sm
import pylab
sm.qqplot(df_sample.request_closing_time,
          line='s')
pylab.show()
```

The above QQplot shows that the data is not normally distributed.

So, to check whether the average response time across complaint type are similar or not, we have to use non parametric test

Krushal Walis H Test

Reject H0: One or more samples are not equal

Fail to reject H0: All sample distribution are equal

```python
complaint_types=df_new['complaint_type'].unique()
complaint_types
```

```python
len(complaint_types)
```

```python
new=[]
for i in complaint_types:
    new.append(df_new.loc[df_new['complaint_type']==i]['request_closing_time'].ravel())
```

```python
#Krushal Walis H Test

from scipy.stats import kruskal
stat,p= kruskal(new[0], new[1], new[2], new[3], new[4], new[5], new[6], new[7], new[8], new[9], new[10], new[11], new[12], new[13], new[14], new[15], new[16], new[17], new[18], new[19], new[20], new[21], new[22])
print(stat, p)
```

```python
if p>0.05:
    print("Fail to reject H0. \nThe mean average response time across complaint type are similar")
else:
    print("reject H0. \nThe mean average response time across complaint type are not similar")
```

 now checking is the type of complaint or service requested or location related
 
 Chi-Square Test of independence
 
 H0: There is no statistically relationship between city and complaint type
 
 H1: There is a statistically relationship between city and complaint type

```python
#contingency table for complaints
cont_table=pd.crosstab(df_new['city'],df_new['complaint_type'], margins=True)
cont_table
```

```python
cont_table.shape
```

```python
observed_values=cont_table.values
```

```python
observed_values
```

```python
#chi-square test

from scipy import stats
ch=stats.chi2_contingency(observed_values)
p_val=ch[1]
```

```python
if p_val<0.05:
    print("Reject the Null Hypothesis. \nThere is a relation between complaint type and location")
else:
    print("Fail to reject the Null Hypothesis. \nThere is no relation between complaint type and location")
```
